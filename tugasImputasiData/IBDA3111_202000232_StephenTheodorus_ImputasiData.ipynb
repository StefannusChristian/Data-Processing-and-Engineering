# Removing Outlier

```python
import numpy as np
def outliers(s, replace=np.nan):
    for i in range(s.shape[1]):
        Q1, Q3 = np.nanquantile(s[i], [0.25 ,0.75])
        IQR = Q3-Q1
        s[i] = s[i].where((s[i] > (Q1 - 1.5 * IQR)) & (s[i] < (Q3 + 1.5 * IQR)), replace)
    return s
```

```python
outlier_removed=outliers(dask_df)
```

```python
for i in range(outlier_removed.shape[1]):
    n_miss = outlier_removed[[i]].isnull().sum()
    perc = n_miss / outlier_removed.shape[0] *100
    print('> %d, Missing: %d (%.1f%%)'%(i,n_miss,perc))
```

```python
from numpy import isnan
from pandas import read_csv
from sklearn.impute import SimpleImputer
```

```python
new_data_stat = outlier_removed
imputer = SimpleImputer(strategy='mean')
imputer.fit(new_data_stat)
trans = imputer.transform(new_data_stat)
print("done")
print('Missing:  %d' % sum(isnan(trans).flatten()))
```
